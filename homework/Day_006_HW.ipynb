# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
print(f'Types of data:\n{app_train.dtypes.value_counts()}')

dtype_selected = app_train.select_dtypes(include = [np.float64, np.int64])

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
for col in dtype_selected:
    if len(np.unique(dtype_selected[col])) == 2:
        del dtype_selected[col]

print("Numbers of remain columns:%d" % dtype_selected.shape[1])

# 檢視這些欄位的數值範圍
n_rows = int(np.ceil(dtype_selected.shape[1] / 2))
fig, axes = plt.subplots(n_rows, 2, figsize=(10, 150))
for (col, ax) in zip(dtype_selected.columns.tolist(), axes.ravel()):
    dtype_selected.boxplot(column = col, ax = ax)
    ax.set_xticklabels([])
    ax.set_title(col)
plt.tight_layout()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE


def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    return data.value_counts().sort_index().cumsum() / len(data)
    
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
cdf = ecdf(app_train['AMT_INCOME_TOTAL'])

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(list(cdf.index), cdf/cdf.max())
plt.xscale('log')
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.show()
```
